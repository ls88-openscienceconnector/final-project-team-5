### Abstract

This project for L&S 88 attempts to recreate the analysis in a study originally published in Palgrave Communications titled [Comparing Apples to Apples](https://www.nature.com/articles/s41599-018-0188-3#Tab1). The purpose of this project is to attempt to reproduce part of the findings in the original analysis regarding the relationship between rain and violent crimes in Boston. We will also evaluate the efficacy and reliability of the methods in the original study to qualify any conclusions that we draw. The original analysis conducted four hypothetical experiments: one for each heat index class ([−18; 0 °C], (0; 24 °C], (24; 35 °C], the Negative, Mild, and High heat index classes, respectively) and one for rainfall. For the purposes of our project, we will replicate just the rainfall hypothetical experiment will the null hypothesis that rainfall decreases daily violent crime count in Boston. 

### Data

The data is from the [Harvard Dataverse](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/4UZ9D4) where is was posted by the researchers of the original study. Research into the data shows that its original provenance is compiled from the following public government databases; [NCDC](https://www.ncdc.noaa.gov/cdo-web), [City of Boston Data](https://data.cityofboston.gov), [City of LA Data](https://data.lacity.org). This data was collected by the city of Boston and the city of LA, respectively. Both data files are CSV files and apppear to have no missing values. Only the Boston data will be used in our reproduction analysis.

### Strategy

The coding analysis will be performed using the `datascience` library. To conduct this portion of our analysis, we will load the data from the CSV file and first explore the data. This looking at the spread of relevant variables and plotting relevant numerical data. Then, we will calculate our test statistic directly from our cleaned data. We then move on to our hypothesis test, which we will conduct by bootstrapping samples from our cleaned data with a for loop and calculating our test statistic from each bootstrapped sample. We then compute a 95% confidence interval for our bootstrapped test statistics and see if our original test statistic value (computed directly from the cleaned data) falls within that confidence interval. We will then come to a conclusion about whether or not to accept our null hypothesis. Finally, in our analysis, we will comment on whether or not the original study was reproducible and what conclusions can be made about our methods compared to the ones used in the study.

### Analysis

The test statistic in this analysis is the **difference between the means of crime counts on rainy and non-rainy days**. That is, for two vectors of rainy day crime counts and non-rainy day crime counts, the test statistic is computed as

$$ {AEE}_i = \frac{1}{N}\sum_{i = 1}^{N}\text{rainy day crime count} - \frac{1}{N}\sum_{i = 1}^{N}\text{non-rainy day crime count} $$

To compute a 95% confidence interval for bootstapped test statistics, the `percentile()` function of the `datascience` package will be used.

To examine the reliability of the results found in the study, we will examine how our simplified methods compare to those of the original, and decide what conclusions can be confidently drawn from our analysis.
